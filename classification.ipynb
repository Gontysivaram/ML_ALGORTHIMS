APPLY THE CAPPING TECHNIQUE TO THE DATA FRAME 

```python
#  capping outliers using IQR method
for col in int_df.columns:
    Q1 = int_df[col].quantile(0.25)
    Q3 = int_df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    int_df[col] = np.where(int_df[col] < lower_bound, lower_bound, int_df[col])
    int_df[col] = np.where(int_df[col] > upper_bound, upper_bound, int_df[col])
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import AdaBoostRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import Lasso, Ridge  # L1 and L2 regularization
```

SPLIT YOUR DATA

```python
target = 'Hazardous'  # Replace with your actual target column

# Split the data into features (X) and target (y)
X = data.drop(columns=[target])  # All features except target
y = data[target]

# Train-test split (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#GENERATE A FUNCTION TO RETREIVE THE METRICS FOR EACH AND EVERY MODEL
```

```python
def evaluate_model(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    # Metrics
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    print(f'Model: {model.__class__.__name__}')
    print(f'Mean Squared Error: {mse}')
    print(f'R2 Score: {r2}')
    print('-' * 50)
```

SUPPORT VECTOR CLASSIFIER

```python
svr = SVR(kernel='rbf')  # RBF kernel works well with non-linear data
evaluate_model(svr, X_train_scaled, y_train, X_test_scaled, y_test)
```

DECISION TREE

```python
dt = DecisionTreeRegressor(random_state=42)
evaluate_model(dt, X_train, y_train, X_test, y_test)
```

```python
#xg boost
```

```python
xgb = XGBRegressor(random_state=42)
evaluate_model(xgb, X_train, y_train, X_test, y_test)
```

```python
#ada boost
```

```python
ada = AdaBoostRegressor(random_state=42)
evaluate_model(ada, X_train, y_train, X_test, y_test)
```

```python
#l1 and l2 for the , regularazitaion
```

```python
lasso = Lasso(alpha=1.0)  # L1 Regularization (alpha is the penalty strength)
evaluate_model(lasso, X_train, y_train, X_test, y_test)
ridge = Ridge(alpha=1.0)  # L2 Regularization
evaluate_model(ridge, X_train, y_train, X_test, y_test)

```

```python
log_reg = LogisticRegression(penalty='l2', solver='lbfgs', max_iter=1000)  # L2 regularization
evaluate_model(log_reg, X_train_scaled, y_train, X_test_scaled, y_test)
```
